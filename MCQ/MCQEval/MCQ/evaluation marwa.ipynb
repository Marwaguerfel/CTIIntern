```python
import pandas as pd

white=pd.read_csv('/content/predicted_answers.csv')
ans=pd.read_csv('/content/cti-mcq.tsv',sep="\t")
```

```python
merged = pd.merge(white, ans, on='Question', suffixes=('_gt', '_model'))
merged
```

```python
def compute_mcq_accuracy(df):
    correct = 0
    total = 0
    for idx, row in df.iterrows():
        pred = row['Predicted Answer']
        gt = row['GT']
        if pred in ['A', 'B', 'C', 'D', 'X']:
            total += 1
        else:
            print('Invalid response at row {}'.format(idx+1))
        if pred == gt:
            correct += 1
    return correct/total*100
```

```python
print('Accuracy:', compute_mcq_accuracy(merged))
```
